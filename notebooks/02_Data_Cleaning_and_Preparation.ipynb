## Missing Value Investigation

The missing values were examined individually to determine whether they represent data-quality issues or valid business scenarios.

### Winner and Player of the Match

All missing values in `winner` and `player_of_match` correspond to matches that ended with "no result".

Since these matches did not produce an official winner, the missing values are expected and will be retained.

### City

Several records contain missing city information despite having valid venue names.

These missing values appear to be incomplete location records rather than valid business events and can be corrected using venue information.

### Conclusion

Most missing values in the dataset are business-related and do not indicate poor data quality.

Only a small number of location records may require correction during data preparation.

## Team Name Consistency Review

The IPL has experienced franchise rebranding and naming changes over the years.

If these variations are not standardized, the same franchise may be treated as multiple teams, resulting in inaccurate performance metrics.

Examples identified:

- Delhi Daredevils → Delhi Capitals
- Kings XI Punjab → Punjab Kings
- Royal Challengers Bangalore → Royal Challengers Bengaluru
- Rising Pune Supergiant → Rising Pune Supergiants

These inconsistencies will be standardized before analysis to ensure franchise-level comparisons remain accurate.

```python
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru',
    'Rising Pune Supergiant': 'Rising Pune Supergiants'
}
```

```python
matches.replace(team_mapping, inplace=True)
deliveries.replace(team_mapping, inplace=True)
```

## Season Format Review

The season field contains inconsistent formats.

Some records use a single-year format (e.g., 2019, 2022), while others use a multi-year format (e.g., 2007/08, 2020/21).

For trend analysis and reporting purposes, a standardized season representation is required.

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches['date'].head()
```

```python
matches['analysis_year'] = matches['date'].dt.year
```

## Venue Name Consistency Review

The venue field contains multiple naming variations for the same stadium.

In several cases, city information has been appended to the venue name, causing the same stadium to appear as separate locations.

Examples include:

- Arun Jaitley Stadium vs Arun Jaitley Stadium, Delhi
- Eden Gardens vs Eden Gardens, Kolkata
- Brabourne Stadium vs Brabourne Stadium, Mumbai

These inconsistencies will be standardized to ensure accurate venue-level analysis.

```python
venue_mapping = {
    'Arun Jaitley Stadium, Delhi': 'Arun Jaitley Stadium',
    'Brabourne Stadium, Mumbai': 'Brabourne Stadium',
    'Barsapara Cricket Stadium, Guwahati': 'Barsapara Cricket Stadium',
    'Dr DY Patil Sports Academy, Mumbai': 'Dr DY Patil Sports Academy',
    'Eden Gardens, Kolkata': 'Eden Gardens',
    'Himachal Pradesh Cricket Association Stadium, Dharamsala': 'Himachal Pradesh Cricket Association Stadium',
    'M Chinnaswamy Stadium, Bengaluru': 'M Chinnaswamy Stadium',
    'MA Chidambaram Stadium, Chennai': 'MA Chidambaram Stadium',
    'Punjab Cricket Association Stadium, Mohali': 'Punjab Cricket Association Stadium',
    'Wankhede Stadium, Mumbai': 'Wankhede Stadium'
}
```

```python
matches['venue'] = matches['venue'].replace(venue_mapping)
```
