# IPL Performance Intelligence: Executive Summary

This project analyzed IPL match and ball-by-ball data to identify patterns associated with franchise success, venue characteristics, scoring behavior, and match outcomes.

The objective was to move beyond descriptive statistics and generate actionable insights that explain how teams achieve success in the IPL.

The analysis covered:

- Match-level performance
- Toss and chasing behavior
- Venue intelligence
- Powerplay and death-over performance
- Win factors
- Franchise benchmarking

# Executive Finding 1

## Winning The Toss Has Limited Impact

Although winning the toss is often viewed as a major advantage, the analysis found that toss winners won approximately 50.6% of matches.

This result suggests that toss outcomes alone do not significantly influence match results and that on-field performance remains the primary driver of success.

# Executive Finding 2

## First Innings Score Is A Major Win Factor

The probability of winning increased consistently with higher first-innings totals.

### Batting First Win Percentage

| Score Band | Win Percentage |
|------------|---------------|
| Below 140 | 19.3% |
| 140–160 | 30.0% |
| 160–180 | 50.0% |
| 180–200 | 60.8% |
| 200+ | 82.6% |

The findings indicate that 180 runs represents an important scoring threshold, while scores above 200 provide a substantial competitive advantage.

# Executive Finding 3

## Powerplay Performance Matters

Winning teams averaged 49.4 powerplay runs compared to 43.7 runs for losing teams.

Teams that established stronger starts were more likely to achieve positive match outcomes, highlighting the importance of maximizing scoring opportunities during the first six overs.

# Executive Finding 4

## Wicket Preservation Is Critical

Winning teams lost an average of 1.10 wickets during the powerplay.

Losing teams lost an average of 1.75 wickets.

The results suggest that successful teams balance aggressive scoring with effective wicket management during the opening phase of an innings.

# Executive Finding 5

## Franchise Success Is Concentrated

Among established franchises:

1. Chennai Super Kings
2. Mumbai Indians
3. Kolkata Knight Riders

recorded the strongest long-term performance metrics.

Chennai Super Kings achieved the highest win percentage and the highest average wins per season, making them the most consistent franchise in the dataset.

# Executive Finding 6

## Different Teams Win In Different Ways

The analysis revealed distinct franchise profiles.

- Rajasthan Royals demonstrated the strongest chasing profile.
- Chennai Super Kings recorded the strongest defending profile.
- Mumbai Indians combined strong chasing and defending capabilities.
- Successful franchises often achieved success through different strategic approaches rather than a single playing style.

# Final Conclusion

The analysis demonstrates that match outcomes are influenced far more by team performance than by pre-match factors such as the toss.

The strongest indicators of success include:

- High first-innings scores
- Strong powerplay performance
- Effective wicket preservation
- Consistent execution across seasons

The findings show that sustained franchise success is built through strategic adaptability, strong batting foundations, and long-term consistency rather than reliance on any single factor.
